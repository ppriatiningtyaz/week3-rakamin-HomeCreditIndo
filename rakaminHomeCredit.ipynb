```python
import pandas as pd
from google.colab import files

dataku = files.upload()
```

```python
import io
df = pd.read_csv(io.BytesIO(dataku['Marketing_Campaign_Data_A-21492758-f47a-4913-971e-68c9a721311a.csv']))

df
```

```python
dataTargetZero = df[df.TARGET == 0]

dataTargetZero
```

```python
dataRegionRating = dataTargetZero[dataTargetZero.REGION_RATING_CLIENT_W_CITY == 3]

dataRegionRating
```

```python
filterDate = dataRegionRating.loc[(dataRegionRating['APPL_DATETIME'] >= '1/1/2017') & (dataRegionRating['APPL_DATETIME'] < '1/1/2019')]

filterDate
```

```python
filterDate['DATE_BIRTH'] = filterDate['DATE_BIRTH'].str.replace("/","-")
filterDate
```

```python
filterDate['DATE_BIRTH'] = pd.to_datetime(filterDate['DATE_BIRTH'], format='%d-%m-%Y')
filterDate
```

```python
filterDate['APPL_DATETIME'] = pd.to_datetime(filterDate['APPL_DATETIME'], format='%d-%m-%Y')
filterDate
```

```python
arrayDate = []
```

```python
# df = df.reset_index(drop=True)
df = filterDate.reset_index(drop=True)
df
```

```python
import numpy as np
for i in range(len(df)):
  umur = df['APPL_DATETIME'][i] - df['DATE_BIRTH'][i]
  umur = umur / np.timedelta64(1, 'Y')
  arrayDate.append(umur)

arrayDate
```

```python
df['AGE'] = arrayDate
df
```

```python
def kategori(num):
  if num >=60:
    return "Above_Sixty"
  if num >=50 :
    return "Fifty"
  if num >= 40:
    return "Fourty"
  if num >= 30:
    return "Thirty"
  if num >= 20:
    return "Twenty"
  
  return "Late Ten"
```

```python
arrayCategory = []

for i in range(len(df)):
  temp = kategori(df['AGE'][i])
  arrayCategory.append(temp)


df['Range Usia'] = arrayCategory
```
